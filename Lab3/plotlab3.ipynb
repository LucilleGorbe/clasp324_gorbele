```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
fac = np.genfromtxt('fac.txt', names=["Accel", "Veloc", "Posit"], delimiter="\t")
cus = np.genfromtxt('cus.txt', names=["Accel", "Veloc", "Posit"], delimiter="\t")
vic = np.genfromtxt('vic.txt', names=["Accel", "Veloc", "Posit"], delimiter="\t")
time = np.genfromtxt('time.txt')
time = time - time[0]
```

```python
figf, axesf = plt.subplots(4, 1, figsize=(6,10))
axesf[0].plot(time, fac["Accel"]/9.80665+1)
axesf[1].plot(time, fac["Accel"])
axesf[2].plot(time, fac["Veloc"])
axesf[3].plot(time, fac["Posit"])

axesf[3].set_xlabel("Time (s)")
figf.suptitle("Factory Calibration, +z Direction")

axesf[0].set_ylabel("Acceleration (G's)")
axesf[0].set_title("Acceleration vs Time")

axesf[1].set_ylabel("Acceleration (m/s^2)")
axesf[1].set_title("Acceleration vs Time, without Gravity")

axesf[2].set_ylabel("Velocity (m/s)")
axesf[2].set_title("Velocity vs Time")

axesf[3].set_ylabel("Altitude (m)")
axesf[3].set_title("Altitude vs Time")

figf.tight_layout()
```

```python
figc, axesc = plt.subplots(4, 1, figsize=(6,10))
axesc[0].plot(time, cus["Accel"]/9.80665+1)
axesc[1].plot(time, cus["Accel"])
axesc[2].plot(time, cus["Veloc"])
axesc[3].plot(time, cus["Posit"])

axesc[3].set_xlabel("Time (s)")
figc.suptitle("Custom Calibration, +z Direction")

axesc[0].set_ylabel("Acceleration (G's)")
axesc[0].set_title("Acceleration vs Time")

axesc[1].set_ylabel("Acceleration (m/s^2)")
axesc[1].set_title("Acceleration vs Time, without Gravity")

axesc[2].set_ylabel("Velocity (m/s)")
axesc[2].set_title("Velocity vs Time")

axesc[3].set_ylabel("Altitude (m)")
axesc[3].set_title("Altitude vs Time")

figc.tight_layout()
```
